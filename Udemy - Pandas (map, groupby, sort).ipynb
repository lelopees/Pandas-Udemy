```python
import pandas as pd
data = pd.read_csv('titanic/train.csv') # Comma Separated Values - Valores Separados Por VÃ­rgula
data.columns = ['IdPassageiro', 'Sobreviveu', 'Classe', 'Nome', 'Sexo', 'Idade', 
                'IrmaoConjuge', 'PaisFilhos', 'Bilhete', 'Tarifa', 'Cabine', 'Embarque']
data.head()
```

### Groupby

```python
display(data.groupby('Sexo').sum())
display(data.groupby('Sexo').mean())
```

```python
data.groupby(['Sexo','Classe', 'Sobreviveu']).mean()
```

```python
data[data['Sobreviveu'] == 1].groupby(['Sexo','Classe']).max()
```

```python
# obter a maior idade por Sexo e Classe dos sobreviventes
data[data['Sobreviveu'] == 1].groupby(['Sexo','Classe']).apply(lambda data: data.loc[data.Idade.idxmax()])
```

```python
data[data['Sobreviveu'] == 1].groupby(['Sexo','Classe'], as_index = False ).mean()
```

```python
data[data['Sobreviveu'] == 1].groupby(['Sexo','Classe']).mean().reset_index()
```

```python
data[data['Sobreviveu'] == 1].groupby(['Sexo','Classe']).agg({'Idade': 'mean', 'PaisFilhos': 'sum'})
```

```python
# obter vÃ¡rias funÃ§Ãµes da idade dos sobreviventes
data[data['Sobreviveu'] == 1].groupby(['Classe']).Idade.agg([min, max])
```
