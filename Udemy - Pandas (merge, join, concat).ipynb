```python
import pandas as pd
 
vendas_df = pd.read_csv(r'Contoso - Vendas  - 2017.csv', sep=';')
produtos_df = pd.read_csv(r'Contoso - Cadastro Produtos.csv', sep=';')  

vendas_df = vendas_df[['Data da Venda', 'Quantidade Vendida', 'ID Produto']] 
produtos_df = produtos_df[['ID Produto', 'Nome do Produto']] 
#usaremos o display para ver todos os dataframes
display(vendas_df)
display(produtos_df) 
```

```python
vendas_df = vendas_df.merge(produtos_df, on='ID Produto')  

display(vendas_df)
```

```python
vendas_df.merge(produtos_df) 
```

```python
vendas_df.merge(produtos_df, on='ID Produto').set_index('Quantidade Vendida')
```

```python
vendas_df.merge(produtos_df, on='ID Produto', how='right')  
```

```python
vendas_df.merge(produtos_df, on='ID Produto', how='outer')  
```

```python
vendas_df.merge(produtos_df, right_index=True, left_index=True)  
```

```python
vendas_df.merge(produtos_df, how='outer', indicator=True)  
```

### JOIN

```python
# rsuffix fuxico da direita
# lsuffix fuxico da esquerda
vendas_df.join(produtos_df, rsuffix='_r', lsuffix='_l')
```

```python
vendas_df.join(produtos_df.set_index('ID Produto'), on ='ID Produto' )
```

```python
vendas_df.join(produtos_df.set_index('ID Produto'), on ='ID Produto' )
```

```python
vendas_df.join(produtos_df.set_index('ID Produto'), on ='ID Produto', how='outer')
```

### Concat

```python
pd.concat([vendas_df,produtos_df], ignore_index=True, join='inner')
```

```python
pd.concat([vendas_df,produtos_df], ignore_index=True, join='inner').drop_duplicates()
```

```python
pd.concat([vendas_df,produtos_df], ignore_index=True, join='inner')
```

```python
pd.concat([vendas_df,produtos_df], keys=['vendas_df','produtos_df'])
```

```python
pd.concat([vendas_df,produtos_df], keys=['vendas_df','produtos_df'], levels=[['vendas_df', 'produtos_df', 'teste']])
```
